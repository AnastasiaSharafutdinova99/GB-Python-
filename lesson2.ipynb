```python
import pandas as pd
```

```python
# 1. Загрузить, просмотреть, определить количество строк и склеить 3 датасета:
# marketing_campaign.csv, users.csv и subscribers.csv.
```

```python
# Загрузка и просмотр: marketing_campaign.csv
df = pd.read_csv('marketing_campaign.csv')
df.head(10)
```

```python
# Количество строк
len(df)
```

```python
# Создание датафрейма из первого датасета для дальнейшего объединения
df_marketing_campaign = df[["user_id", "date_served", "marketing_channel", "variant", "language_displayed", "converted"]]
print(df_marketing_campaign.shape)
df_marketing_campaign.head(10)
```

```python
# Загрузка и просмотр: users.csv
df = pd.read_csv('users.csv')
df.head(10)
```

```python
# Количество строк
len(df)
```

```python
# Создание датафрейма из второго датасета для дальнейшего объединения
df_users = df[["user_id", "age_group", "language_preferred"]]
print(df_users.shape)
df_users.head(10)
```

```python
# Загрузка и просмотр: subscribers.csv
df = pd.read_csv('subscribers.csv')
df.head(10)
```

```python
# Количество строк
len(df)
```

```python
# Создание датафрейма из третьего датасета для дальнейшего объединения
df_subscribers = df[["user_id", "subscribing_channel", "date_subscribed", "date_canceled", "is_retained"]]
print(df_subscribers.shape)
df_subscribers.head(10)
```

```python
# Объединение первых двух датасетов
df_marketing_campaign_users = df_marketing_campaign.merge(df_users, left_on = ["user_id"], right_on = ["user_id"], how = "left")
df_marketing_campaign_users
```

```python
# Объединение трёх датасетов
df_marketing_campaign_users_subscribers = df_marketing_campaign_users.merge(df_subscribers, left_on = ["user_id"], right_on = ["user_id"], how = "left")
df_marketing_campaign_users_subscribers
```

```python
# Количество
df_marketing_campaign_users_subscribers.shape
```

```python
# Типы
df_marketing_campaign_users_subscribers.info()
```

```python
# Описательная статистика
df_marketing_campaign_users_subscribers.describe()
```

```python
# 3. Определить эффективность маркетинговых каналов по привлечению платящих игроков.
```
